```python
import torch

# 假设这是你的 Tensor，且它在计算图中
t = torch.rand((3, 3, 3, 3), dtype=torch.float32, requires_grad=True)  # 示例 Tensor，跟踪梯度

# 将 Tensor 从计算图中分离，然后转换为 NumPy 数组
P = t.detach().numpy()

# 如果 Tensor 在 GPU 上，先将其移动到 CPU，再转换为 NumPy 数组
# P = t.detach().cpu().numpy()

# 打印 NumPy 数组 P
print(P)  # 输出类似于 array([...], dtype=float32)

```
